# **#05 Membalik Urutan Baris dan Kolom pada Data Frame**

Import Modules

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan Data Frame

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')
# Nilai random antara 1 - 10 dengan dimensi yang ditentukan oleh n_rows dan n_cols

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns=cols)
df
```

Membalik Urutan Kolom

```python
df.loc[:, ::-1] # Fungsi loc[] memiliki dua parameter yang berfungsi untuk mengatur baris dan kolom.
```

Membalik Urutan Baris

```python
df.loc[::-1] # Untuk membalik urutan baris, ubah stepper pada ruas pertama menjadi -1
```

Membalik Urutan Baris dan Melakukan Penyesuaian Ulang Index

```python
df.loc[::-1].reset_index(drop=True) # Jika kita ingin membalik urutan baris, tetapi index 0 tetap diatas, kita bisa menggunakan reset_index().
```

# **#06 Mengganti Nama (label) Kolom pada Data Frame**

Import Modules

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan Data Frame

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')
# Nilai random antara 1 - 10 dengan dimensi yang ditentukan oleh n_rows dan n_cols

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns=cols)
df
```

Mengganti Nama (label) Untuk Sebuah Kolom pada Data Frame

```python
df.rename(columns={'C':'Hobi'}) 
# Fungsi rename akan mengganti nama sebuah kolom pada data frame dengan 'c' yang merupakan kolom targer dan 'hobi' sebagai nama pengganti kolom dengan nama 'c'.
```

Mengganti Nama (label) Untuk Banyak Kolom pada Data Frame

```python
df.rename(columns={'A':'Nama', 'B':'Alamat', 'D':'Kota'}) 
# Menggunakan cara yang sama seperti sebelumnya, namun digunakan dengan beberapa pasangan keys dan values.
```

# **#07 Menghapus (drop) Missing Values (NaN) pada Data Frame**

Import Modules

```python
import pandas as pd

print(pd.__version__)
```

Persiapan Data Frame

```python
df = pd.util.testing.makeMissingDataframe().reset_index() 
df.head()
# Fungsi pd.util.testing.makeMissingDataframe() digunakan untuk menghasilkan data dummy yang memiliki nilai NaN (Not a Number).
# Fungsi reset_index() berfungsi untuk membentuk index baru, menggantikan kolom index yang lama.
# Fungsi head() berfungsi untuk menampilkan lima data awal.
```

```python
df = df.rename(columns={'index':'Z'})
df.head()
# Fungsi rename() digunakan untuk mengubah label kolom 'index' (yang lama) menjadi 'Z'
```

```python
df_backup = df.copy(deep=True)
# Mencadangkan nilai df ke variabel df_backup menggunakan fungsi copy().
```

Menghapus (drop) Setiap Kolom yang Mengandung Missing Values

```python
df = df.dropna(axis='columns') 
df.head() 
# Menggunakan fungsi dropna() yang diberi parameter axis dengan value 'columns'.
```

Menghapus (drop) Setiap Baris yang Mengandung Missing Values

```python
df = df_backup.copy(deep=True)

df = df.dropna(axis='rows') 
# Menggunakan fungsi dropna() yang diberi parameter axis dengan value 'rows'.
df.head()
```

Persentase missing values untuk tiap kolom

```python
df = df_backup.copy(deep=True)

df.isna().mean()*100
# Fungsi isna() akan mengumpulkan semua missing value untuk setiap kolom dari data frame.
# Fungsi mean() akan menghitung rata-rata dari missing value yang dimiliki setiap kolom.
```

Menghapus (drop) Setiap Kolom yang Mengandung Missing Values Berdasarkan Threshold (ambang batas)

```python
treshold = len(df) * 0.9 # len() digunakan untuk menghitung panjang baris dan variable thershold yang berisi batas 0.9 atau 90%

df = df.dropna(thresh=treshold, axis='columns')
df.head()
```

# **#08 Memeriksa Kesamaan Antar Dua Buah Kolom (Series) pada Data Frame**

Import Modules

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan Data Frame

```python
# Membuat dua kolom yang memiliki nilai yang sama
data = {'A':[15, 15, 18, np.nan, 12], 
        'B':[15, 15, 18, np.nan, 12]}

df = pd.DataFrame(data)
df
```

Mengenal Pandas Series
- Series adalah struktur data lain yang dimiliki pandas dengan objek 1-dimensi yang berisi sequence nilai dan berasosiasi dengan label data, yang disebut indeks.
- Untuk membuat sebuah Series atau kolom, kita bisa membuatnya dari sebuah array dengan memanggil fungsi Series pada pandas.


```python
df['A']
```

```python
type(df['A'])
```

```python
type(df)
```

Memeriksa Kesamaan dengan Operator ==

```python
df['A'] == df['B'] # operator == tidak dapat membandingkan nilai NaN dengan baik.
```

Memeriksa kesamaan dengan method equals()

```python
df['A'].equals(df['B'])
```

Memeriksa Kesamaan Antar Dua Data Frame

```python
df1 = df.copy(deep=True)

df.equals(df1)
```
