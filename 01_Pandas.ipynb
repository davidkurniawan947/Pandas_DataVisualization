# **#02: Pemilihan baris (rows selection) pada Data Frame**

# **Persiapan Data Frame**

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), 
                  columns = cols) #Jangkauan angka 1-10 dengan dimensi yang ditentukan oleh n_rows dan n_cols
df
```

# **Selection dengan operator logika |**

```python
df[(df['A'] == 1) | (df['A'] == 3)]
```

# Selection dengan fungsi **isin()**

```python
df[df['A'].isin([1, 3])] #logika di balik fungsi isin() sama dengan logika OR |
```

# **Mengenal Operator negasi ~**

operator negasi digunakan untuk mengenali not

```python
df[-df['A'].isin([1, 3])]
```

# **#03 Konversi tipe data String ke Numerik pada kolom Data Frame**

# **Persiapan Data Frame**

```python
data = {'col1':['1', '2', '3', 'teks'],
        'col2':['1', '2', '3', '4']} #2 keys yaitu col1 & col2 dan akan menjadi nama kolom
        #values yang pertama akan berisikan data-data yang berada pada kolom pertama
        
df = pd.DataFrame(data)
df
```

```python
df.dtypes #digunakan untuk menecek type data
```

# Konversi tipe data dengan fungsi **astype()** 

```python
df_x = df.astype({'col2':'int'})
df_x
```

```python
df_x.dtypes
```

# Konversi tipe data numerik dengan fungsi **to_numeric()**

```python
df.apply(pd.to_numeric, errors='coerce')
```

# **#4 Pemilihan kolom (columns selection) pada Data Frame berdasarkan tipe data**

# **Persiapan Data Frame**

```python
n_rows = 5
n_cols = 2
cols = ['bil_pecahan', 'bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df['bil_pecahan'] = df['bil_pecahan'].astype('float') #menkonversi pecahan menjadi tipe data float menggunakan function astype

df.index = pd.util.testing.makeDateIndex(n_rows, freq= 'H') #jumlah data disesuaikan dengan n_rows, data-data yang di generate memiliki interval jam yaitu 1 jam sekali
df = df.reset_index()

df['teks'] = list('ABCDE')

df
```

```python
df.dtypes
```

# **Memilih kolom bertipe data numerik**

```python
df.select_dtypes(include='number') #fungsi ini akan memanggil kolom yang bertipe data numerik saja
```

```python
df.select_dtypes(include='float') #fungsi ini akan memanggil kolom yang bertipe data float saja
```

```python
df.select_dtypes(include='int') #fungsi ini akan memanggil kolom yang bertipe data int saja
```

# Memilih kolom bertipe data string atau **object**

```python
df.select_dtypes(include='object') #fungsi ini akan memanggil kolom yang bertipe data numerik saja
```

# Memilih kolom bertipe data string atau **datetime**

```python
df.select_dtypes(include='datetime') #fungsi ini akan memanggil kolom yang bertipe data date saja
```

# Memilih kolom dengan kombinasi tipe data

```python
df.select_dtypes(include=['number', 'object']) #fungsi ini akan memanggil kolom yang memiliki tipe data numerik dan string
```
